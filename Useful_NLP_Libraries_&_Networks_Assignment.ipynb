**Question 4: Describe the architecture and functioning of a Recurrent Natural Network (RNN).**

A Recurrent Neural Network (RNN) is a type of deep learning architecture specifically built for sequential data, such as text, audio, or stock prices.

Unlike standard networks that process inputs independently, an RNN uses a looping mechanism that allows information to persist. At each step in a sequence, the network takes the current input and combines it with a "hidden state" (memory) representing previous inputs. This creates a chain-like structure that enables the model to understand context—for example, predicting the next word in a sentence based on the words that came before it.

However, standard RNNs often struggle with long sequences due to the vanishing gradient problem, where early information is lost over time.


**Question 5: What is the key difference between LSTM and GRU networks in NLP applications?**

The key difference between Long Short-Term Memory (LSTM) and Gated Recurrent Unit (GRU) networks in NLP applications lies in their architectural complexity and memory management.

LSTM is more complex, utilizing three gates (input, forget, and output) and maintaining two separate states (cell state and hidden state).

GRU is a simplified, more efficient variant, using only two gates (reset and update) and one hidden state, allowing for faster training.

**Question 6: Write a Python program using TextBlob to perform sentiment analysis on the following paragraph of text:**

“I had a great experience using the new mobile banking app. The interface is intuitive, and customer support was quick to resolve my issue. However, the app did crash once during a transaction, which was frustrating"

**Your program should print out the polarity and subjectivity scores.**

```python
from textblob import TextBlob

# The input text
text_paragraph = """
I had a great experience using the new mobile banking app. The interface is intuitive,
and customer support was quick to resolve my issue. However, the app did crash once
during a transaction, which was frustrating.
"""

# Creating a TextBlob object
blob = TextBlob(text_paragraph)

# Extracting sentiment scores
polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

# Printing the results
print(f"Sentiment Analysis Results:")
print(f"---")
print(f"Polarity: {polarity:.3f}")
print(f"Subjectivity: {subjectivity:.3f}")
```

**Question 7: Given the sample paragraph below, perform string tokenization and frequency distribution using Python and NLTK:**

“Natural Language Processing (NLP) is a fascinating field that combines linguistics, computer science, and artificial intelligence. It enables machines to understand, interpret, and generate human language. Applications of NLP include chatbots, sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.”

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Ensure the necessary resource is downloaded
nltk.download('punkt')

# The sample paragraph
text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical."""

# 1. String Tokenization
tokens = word_tokenize(text.lower())  # Converting to lower case for accurate counting

# 2. Frequency Distribution
fdist = FreqDist(tokens)

# Displaying Results
print("Tokens:", tokens)
print("\nFrequency Distribution:")
for word, frequency in fdist.most_common(10):
    print(f"'{word}': {frequency}")
```

**Question 8: Implement a basic LSTM model in Keras for a text classification task using the following dummy dataset. Your model should classify sentences as either positive (1) or negative (0).**

-Dataset

texts = [
“I love this project”, #Positive
“This is an amazing experience”, #Positive
“I hate waiting in line”, #Negative
“This is the worst service”, #Negative
“Absolutely fantastic!” #Positive
]

labels = [1, 1, 0, 0, 1]

**Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on this data. You may use Keras with TensorFlow backend.**

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# 1. Dataset Preparation
texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic!"
]
labels = np.array([1, 1, 0, 0, 1])

# 2. Preprocessing & Tokenization
max_words = 1000
tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# Padding sequences so they all have the same length
max_len = 10
data = pad_sequences(sequences, maxlen=max_len)

# 3. Building the LSTM Model
model = Sequential([
    # Embedding layer: turns word indices into dense vectors of fixed size
    Embedding(input_dim=max_words, output_dim=16, input_length=max_len),
    # LSTM layer: processes the sequence of vectors
    LSTM(32),
    # Output layer: Sigmoid activation for binary classification (0 or 1)
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 4. Training the Model
# Note: With only 5 samples, this is purely for demonstration
model.fit(data, labels, epochs=10, verbose=0)

print("Model training complete.")
print("Sample prediction for 'I love this project':", model.predict(data[0:1]))
```

**Question 9: Using spaCy, build a simple NLP pipeline that includes tokenization, lemmatization, and entity recognition. Use the following paragraph as your dataset:**

“Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the development of India’s atomic energy program. He was the founding director of the Tata Institute of Fundamental Research (TIFR) and was instrumental in establishing the Atomic Energy Commission of India.”

**Write a Python program that processes this text using spaCy, then prints tokens, their lemmas, and any named entities found.**

```python
import spacy

# 1. Load the English language model
# Make sure to run 'python -m spacy download en_core_web_sm' in your terminal first
nlp = spacy.load("en_core_web_sm")

# 2. Dataset
text = ("Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the "
        "development of India’s atomic energy program. He was the founding director of the "
        "Tata Institute of Fundamental Research (TIFR) and was instrumental in establishing "
        "the Atomic Energy Commission of India.")

# 3. Process the text through the pipeline
doc = nlp(text)

# 4. Tokenization and Lemmatization
print(f"{'Token':<20} | {'Lemma':<20}")
print("-" * 42)
for token in doc[:10]:  # Displaying first 10 tokens for brevity
    print(f"{token.text:<20} | {token.lemma_:<20}")

# 5. Named Entity Recognition (NER)
print("\nNamed Entities:")
print("-" * 42)
for ent in doc.ents:
    print(f"{ent.text:<35} | {ent.label_}")
```
